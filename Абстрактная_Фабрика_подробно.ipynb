## Абстрактная Фабрика


Абстрактная фабрика определяет метод создания для каждого продукта. Каждый фабричный метод инкапсулирует новый оператор и конкретные, специфичные для платформы классы продуктов. Затем каждая «платформа» моделируется классом, производным от Factory.


image.png

Цель абстрактной фабрики — предоставить интерфейс для создания семейств связанных объектов без указания конкретных классов. Этот рисунок встречается в оборудовании для штамповки листового металла, используемом при производстве японских автомобилей. Штамповочное оборудование представляет собой абстрактную фабрику, которая создает детали кузова автомобиля. Одно и то же оборудование используется для штамповки правых дверей, левых дверей, правых передних крыльев, левых передних крыльев, капотов и т. д. для разных моделей автомобилей. Благодаря использованию роликов для замены штампов, конкретные модели, производимых машин, можно изменить в течение трех минут.

image.png

```python
import abc


class AbstractFactory(metaclass=abc.ABCMeta):
    """
    Создаем шаблон Абстрактной фабрики
    """

    @abc.abstractmethod
    def create_product_a(self):
        pass

    @abc.abstractmethod
    def create_product_b(self):
        pass


class ConcreteFactory1(AbstractFactory):
    """
    Создаем первую реальную фабрику по шаблону
    """

    def create_product_a(self):
        return ConcreteProductA1()

    def create_product_b(self):
        return ConcreteProductB1()


class ConcreteFactory2(AbstractFactory):
    """
    Создаем вторую реальную фабрику по шаблону
    """

    def create_product_a(self):
        return ConcreteProductA2()

    def create_product_b(self):
        return ConcreteProductB2()


class AbstractProductA(metaclass=abc.ABCMeta):
    """
    Создаем шаблон для продукта, который хотим получать из фабрики
    """

    @abc.abstractmethod
    def interface_a(self):
        pass


class ConcreteProductA1(AbstractProductA):
    """
    Создаем класс конкретного продукта типа А, который требует бизнес, 
    со своим функционалом
    """

    def interface_a(self):
        pass


class ConcreteProductA2(AbstractProductA):
    """
    Создаем класс второго продукта типа А, который требует бизнес, 
    со своим функционалом
    """

    def interface_a(self):
        pass


class AbstractProductB(metaclass=abc.ABCMeta):
    """
    Создаем абстрактный продукт (шаблон продуктов)
    """

    @abc.abstractmethod
    def interface_b(self):
        pass


class ConcreteProductB1(AbstractProductB):
    """
    Создаем класс конкретного продукта типа B, который требует бизнес, 
    со своим функционалом
    """

    def interface_b(self):
        pass


class ConcreteProductB2(AbstractProductB):
    """
    Создаем класс второго продукта типа B, который требует бизнес, 
    со своим функционалом
    """

    def interface_b(self):
        pass


def main():
    for factory in (ConcreteFactory1(), ConcreteFactory2()):
      # каждая фабрика создает оба продукта
        product_a = factory.create_product_a() 
        product_b = factory.create_product_b()
      # каждый продукт может использовать свои методы
        product_a.interface_a()
        product_b.interface_b()
      # ВАЖНО продукты, произведенные на разных фабриках могут имень различные значения аттрибутов 


if __name__ == "__main__":
    main()
```
